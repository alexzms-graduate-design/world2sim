```python
import mph

client = mph.start()
model = client.load("./transient_real_pressure.mph")
mph.tree(model)
print(model.parameters())
```

```python
# modify the model parameters(global)
```

```python
T_end = 0.1
frequencies = [50, 75, 100, 125, 150, 175, 200, 225, 250]
Z_bound = [3000, 3250, 3500, 3750, 4000, 4250, 4500, 4750, 5000, 5250, 5500, 5750, 6000]
for i in range(len(frequencies)):
    for j in range(len(Z_bound)):
        model.parameter("Tend", T_end)
        model.parameter("f0", frequencies[i])
        model.parameter("Z_bound", Z_bound[j])
        model.build()
        model.solve("研究 1")
        save_file_name = f"python/outputs/output_f={frequencies[i]}_Z={Z_bound[j]}_T={T_end}.csv"
        model.export("plot1", save_file_name)
        print("Exported to", save_file_name)
```
